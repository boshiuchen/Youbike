# Date fix test code

## Solve by conversion to datetime, and plus 1 day

```python
匯入套件
import pandas as pd
from datetime import timedelta
from pathlib import Path 
```

```python
# 匯入要改的資料
df_weather = pd.read_csv('./HistoricalWeather2017-2022.csv')
```

```python
df_weather.tail()
```

```python
# 把原本24:00 -> 00:00 的時間改回 24:00
# change 01:00:00 back to 24:00:00
df_weather["NewHour"]=df_weather["NewHour"].apply(lambda x: '24:00:00' if x=='00:00:00'  else x)
df_weather
```

```python
# 匯入新資料
df_weatherMakeup = pd.read_csv('./HistoricalWeathermakeupfirst2016.csv')
```

```python
df_weatherMakeup.tail()
```

```python
# 結合資料
df_weathernew=pd.concat([df_weather, df_weatherMakeup],ignore_index=True)
df_weathernew
```

```python
# 去掉不需要的
df_weathernew = df_weathernew.drop(['Unnamed: 0','Date','ObsTime'],axis=1)
df_weathernew.head()
```

```python
df_weathernew
```

```python
df_weathernew=df_weathernew.sort_values(by=['StationID','NewDate', 'NewHour'])
```

```python
df_weathernew.head(25)
```

## Start the conversion
### Mask
### Convert 24 ->0
### for the converted cases, add 1 day

```python
mask = df_weathernew['NewHour'] == "24:00:00"
```

```python
df_weathernew['NewDate']=pd.to_datetime(df_weathernew['NewDate'])
```

```python
df_weathernew
```

```python
df_weathernew.loc[mask, 'NewHour']='00:00:00'
df_weathernew
```

```python
df_weathernew['NewDate']
```

```python

df_weathernew.loc[mask, 'NewDate']=df_weathernew['NewDate']+timedelta(days=1)
df_weathernew
```

```python
df_weathernew
```

```python
filepath = Path('./Weather.csv')  
# filepath.parent.mkdir(parents=True, exist_ok=True)  
df_weathernew.to_csv(filepath,index=False)  
```
