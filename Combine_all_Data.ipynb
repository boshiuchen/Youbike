```python

# read csv file
years = [2017,2018,2019,2020,2021,2022]
resultMain = pd.DataFrame()
resultMakeup = pd.DataFrame()
result = pd.DataFrame()

# Read in main data
for year in years:
    tmpMain = pd.read_csv(f'HistoricalWeather{year}.csv',low_memory=False)
    resultMain=pd.concat([resultMain,tmpMain],ignore_index=True)
    
# Read in old data
for year in years:
    tmp = pd.read_csv(f'HistoricalWeathermakeup{year}.csv',low_memory=False)
    resultMakeup=pd.concat([resultMakeup,tmp],ignore_index=True)

    
result=pd.concat([resultMain,resultMakeup],ignore_index=True)
```
