```python
import pandas as pd
import os
import glob
import numpy as np
```

```python
path = r"./"
all_files = glob.glob(os.path.join(path, "*日曆表.csv"))
```

```python
all_files
```

```python
li = []

for file in all_files:
    df = pd.read_csv(file,encoding='cp950',index_col=None)
    li.append(df)
```

```python
df = pd.concat(li, axis=0, ignore_index=True)
```

```python
df
```

```python
df = df.drop_duplicates()
df['date'] = pd.to_datetime(df['西元日期'], format='%Y%m%d')
df['是否放假'] = df['是否放假']/2
```

```python
df['weekday'] = np.where(df['星期']=='日', 'Sunday',
             np.where(df['星期']=='六', 'Saturday',
             np.where(df['星期']=='五', 'Friday',
             np.where(df['星期']=='四', 'Thursday',
             np.where(df['星期']=='三', 'Wednesday',
             np.where(df['星期']=='二', 'Tuesday',
              'Monday'))))))
```

```python
#df.drop('西元日期',axis=1, inplace=True)
df.rename(columns={ '是否放假': 'holiday', '備註' :'notes'}, inplace=True)
```

```python
df
```

```python
df = df[ [ 'date','weekday','holiday','notes' ] ]
```

```python
df.head()
```

```python
df.to_csv('./Calendar.csv',index=False,header=True)
```
