```python
import pandas as pd
import os
import glob
import numpy as np
```

```python
# https://ithelp.ithome.com.tw/articles/10262521

path1 = r"./TaipeiCity/YouBike_All_History/YouBike1"

all_files1 = glob.glob(os.path.join(path1, "*票證刷卡資料.csv"))
```

```python
# 確認一下有哪些檔案

all_files1
```

```python
# https://stackoverflow.com/questions/20906474/import-multiple-csv-files-into-pandas-and-concatenate-into-one-dataframe

# youbike有些csv檔案有header，有些沒有，因此會造成union檔案時發生格是怪怪的問題
# 解決辦法 header=None then drop 'header-like' rows

# for 2019,2020,2022 set header=None

# for 202102 notepad++開啟檔案 複製出一份正常的csv
df1 = pd.concat((pd.read_csv(f,encoding='utf-8', header=None, on_bad_lines='warn') for f in all_files1), ignore_index=True)

df1.columns = ['rent_time','rent_station','return_time','return_station','rent','infodate']

# 新增Youbike代數欄位
df1['Generation']=1
```

```python
path2 = r"./TaipeiCity/YouBike_All_History/YouBike2"

all_files2 = glob.glob(os.path.join(path2, "*票證刷卡資料.csv"))

# there are headers for youbike2 
df2 = pd.concat((pd.read_csv(f,encoding='utf-8', on_bad_lines='warn') for f in all_files2), ignore_index=True)

df2['Generation']=2
```

#### 整理MRT進出站匯總人數

```python
# https://ithelp.ithome.com.tw/articles/10262521

MRT_path = r"./捷運進出站人數資料"

MRT_files = glob.glob(os.path.join(MRT_path, "*.csv"))

MRT_files
```

```python
# union個別檔案成為一個df

MRT_df = pd.concat((pd.read_csv(f,encoding='utf-8') for f in MRT_files), ignore_index=True)
```

```python
# 年, 月, 日 to date 

# https://stackoverflow.com/questions/19350806/how-to-convert-columns-into-one-datetime-column-in-pandas

MRT_df['date'] = pd.to_datetime(dict(year=MRT_df.年, month=MRT_df.月, day=MRT_df.日))
```

```python
# https://stackoverflow.com/questions/23836277/add-leading-zeros-to-strings-in-pandas-dataframe

MRT_df['時段'] = MRT_df['時段'].astype(str).apply(lambda x: x.zfill(2))
```

```python
MRT_df['time_period'] = MRT_df['時段'].apply(lambda x: f'{x}:00:00')
```

```python
# 丟掉 年, 月, 日, 時 選擇 date, time_period 

MRT_df = MRT_df.iloc[:,[0,1,6,7,8]]
```

```python
# 欄位重新命名

MRT_df.rename(columns = {'車站編號':'MRT_No','出站':'MRT_Station','人次':'MRT_Out_ppl'}, inplace = True)
MRT_df.head()
```

```python
MRT_df.to_csv('MRT_All.csv')
```

### 先切出一小部分YouBike資料整理

```python
# 隨機選 才會1代2代都有

import random

random.seed(888)
```

```python
df_100K = df.sample(100000)
```

```python
df_100K_copy = df_100K.copy()
```

```python
# 拆出rent日期跟時間

df_100K_copy[['rent_date','rent_time_period']] = df_100K_copy['rent_time'].str.split(' ', -1, expand=True)
```

```python
# 拆出return日期跟時間

df_100K_copy[['return_date','return_time_period']] = df_100K_copy['return_time'].str.split(' ', -1, expand=True)
```

```python
# 把 一個字串 時:分:秒 拆成 3個整數 時 分 秒 欄位

df_100K_copy[['rent_hr','rent_min','rent_sec']] = df_100K_copy['rent'].str.split(':', -1, expand=True).astype(int)
```

```python
# 換算成秒數

df_100K_copy['rent(sec)'] = 3600 * df_100K_copy['rent_hr'] + 60 * df_100K_copy['rent_min'] + df_100K_copy['rent_sec']
```

```python
df_100K_rent = df_100K_copy.iloc[:,[0,1,4,5,6,7,13]]

df_100K_rent.head()
```

```python
# 參考上面的rent, return也切一份出來

df_100K_return = df_100K_copy.iloc[:,[2,3,4,5,8,9,13]]

df_100K_return.head()
```

### 準備group by rent

```python
rent_group = df_100K_rent.groupby(['rent_station','rent_time','rent_date','rent_time_period','Generation'])
```

```python
df_100K_rent_output = rent_group['rent(sec)'].aggregate(['count','sum'])
```

```python
df_100K_rent_output.head()
```

```python
# rename coulmns
df_100K_rent_output.rename(columns={'count':'rent_count', 'sum':'rent(sec)'}, inplace=True)
```

```python
df_100K_rent_output.head()
```

### 準備group by return

```python
return_group = df_100K_return.groupby(['return_station','return_time','return_date','return_time_period','Generation'], as_index=False)
```

```python
df_100K_return_output = return_group['rent(sec)'].aggregate(['count','sum'], as_index=False)
```

```python
df_100K_return_output.rename(columns={'count':'return_count', 'sum':'rent(sec)'}, inplace=True)
```

```python
#df_100K_return_output.head()
```
