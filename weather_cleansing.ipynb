```python
# 石牌測站 C0AI40 繼承 C0A9B0
# 捨棄 2019/09 以後的 C0A9B0
year = ['2020','2021','2022']
for i in year:
    bdf1 = df[(df.yy == i) & (df.StationID == 'C0A9B0')].index
    df.drop(bdf1, inplace=True)
    
month = ['09','10','11','12']
for m in month:
    bd = df[(df.yy == '2019')&(df.mm == m)&(df.StationID == 'C0A9B0')].index
    df.drop(bd, inplace = True)
```

```python
# df['ObsTime'] = pd.to_numeric(df['ObsTime'], errors='coerce') # 轉換 ObsTime 為數值型態
data = df[(df.Precp == 'X')] # 將雨量值為 X 的欄位存入另一個變數，裡面全部都是 X
dd = data['NewDate'].tolist() # 將 data 裡的 Date 依序取出製作成 list
do = data['ObsTime'].tolist() # 將 data 裡的 ObsTime 依序取出製作成 list

t = np.arange(0,88) # 雨量缺失值 X 共 88 筆資料
for i in t:
    mask1 = (df.NewDate == dd[i])
    mask2 = (df.ObsTime == do[i])
    mask3 = (df.Precp != 'X') & (df.Precp != '...') & (df.Precp != '/')
    nadf = df[mask1 & mask2 & mask3]
    av_pre = pd.to_numeric(nadf.Precp).sum()/(nadf.Precp.count()) # 先將所有有效測站該時段的值轉換為數字型態，進行加總，算出平均數
    round(av_pre, 1) # 將該平均數進位到小數點後第一位
    df.loc[mask1 & mask2 & (df.Precp == 'X'), 'Precp'] = round(av_pre, 1) # 填補缺失值
```

```python
dataTp = df[(df.WS == '...')]
dd = dataTp['NewDate'].tolist()
do = dataTp['ObsTime'].tolist()
# print(dd[47676])
# print(do[47676])
# dataTp['Temperature'].info()

t = np.arange(0,50330)
for i in t:
    mask1 = (df.NewDate == dd[i])
    mask2 = (df.ObsTime == do[i])
    mask3 = (df.WS != '...') & (df.WS != '/') & (df.WS != 'X')
    nadf = df[mask1 & mask2 & mask3]
    temp = nadf.loc[:,'WS']
    temp = pd.to_numeric(temp, errors='coerce') # 轉數值型態才不會報錯(非數值型態無法做排序)
    array = np.array(temp) # 將不是 '...' 的溫度欄位做成陣列
    vals,counts = np.unique(array, return_counts=True) # 將唯一值儲存在 vals 陣列中
    index = np.argmax(counts)
    if index == 0: # 代表沒有重複的值可以做為眾數
        av_pre = pd.to_numeric(nadf.WS).sum()/(nadf.WS.count()) # 先將所有有效測站該時段的值轉換為數字型態，進行加總，算出平均數
        round(av_pre, 1) # 將該平均數進位到小數點後第一位
        df.loc[mask1 & mask2 & (df.WS == '...'), 'WS'] = round(av_pre, 1) # 以平均數填補缺失值
    else:
        df.loc[mask1 & mask2 & (df.WS == '...'), 'WS'] = vals[index] # 以眾數填補缺失值
```
